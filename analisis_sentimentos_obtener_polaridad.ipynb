```python
import pandas as pd
import numpy as np
from transformers import pipeline
#libreria para expreciones regulares
import re
```

```python
df = pd.read_excel("/content/tweetspresidenciaperu2023.xlsx")
```

```python
df.head()
```

```python
#creo un nuevo array con la informacion de la columna text
data=df[['text']]
```

```python
print(data)
```

```python
#Mostrar los 5 primeros registros de mi array
data.head()
```

```python
def cleaning_text(texto):

    texto = re.sub(r'http\S+', '', texto)
    texto = re.sub(r'@\w+', '', texto)
    texto = re.sub(r'#\w+', '', texto)
    texto = re.sub(r'[^\w\s\\]', '', texto)
    texto = re.sub(r'\d+', '', texto)
    texto = re.sub(r'\r?\n', ' ', texto)
    texto = texto.lower()
    return texto

data['text'] = data['text'].apply(cleaning_text)
```

```python
#Convertir el array data a una lista de string
text_list = data['text'].tolist()
```

```python
#obtener el total de item en mi lista
print(len(text_list))
```

```python
#Creo un modelo nuevo para poder obtener la polaridad de los textos
#refiniteautomata/beto-sentiment-analysis este modelo es un modelo preentrenado que permite obtener la polaridad de los textos: triste, alegre o neutral
specific_model = pipeline(model="finiteautomata/beto-sentiment-analysis")
```

```python
#Realizo la prediccion de los textos para obtener la polaridad: negativo, positivo, neutral
text_results = specific_model(text_list)
```

```python
#creao un nuevo dataFrame con el resultado obtenido
polarity = pd.DataFrame(text_results)
```

```python
# Relaciono los textos con su respectivo polaridad
data = data.join(polarity)
```

```python
data.head()
```

```python
#remplazo cada textos segun su polaridad: neutral: 0, Negativo: -1, Positovo: 1
def create_column_target(data):
    mapping = {'NEU': 0, 'NEG': -1, 'POS': 1}
    data['target'] = data['label'].map(mapping)
    return data

data = create_column_target(data)
```
